### Cosine distance


```python
x=["D1","D2","D3","D4"]
y_xr_auc=df_xray["AUC_mean"].values[8:]
y_ct_auc=df_ct["AUC_mean"].values[8:]
y_xr_acc=df_xray["Acc_mean"].values[8:]
y_ct_acc=df_ct["Acc_mean"].values[8:]
y_xr_spec=df_xray["Spec_mean"].values[8:]
y_ct_spec=df_ct["Spec_mean"].values[8:]
y_xr_sen=df_xray["Sen_mean"].values[8:]
y_ct_sen=df_ct["Sen_mean"].values[8:]

y_xr_dunn=df_xray["Dunn_mean"].values[8:]
y_ct_dunn=df_ct["Dunn_mean"].values[8:]
y_xr_dav=df_xray["Davies_mean"].values[8:]
y_ct_dav=df_ct["Davies_mean"].values[8:]
y_xr_sil=df_xray["Sil_mean"].values[8:]
y_ct_sil=df_ct["Sil_mean"].values[8:]
```

```python
plt.rcParams['figure.dpi'] = 800
plt.rcParams['savefig.dpi'] = 800
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
```

```python
plt.vlines(y=x, xmin = 0,color='skyblue')
plt.plot(y_xr_acc, marker = "o", linewidth = 5)
```

```python
sns.set_style('darkgrid')
width=0.5
fig,ax=plt.subplots(1,4,figsize=(14,3))
x_axis= np.arange(len(x))
_,stems_1,_=ax[0].stem(x_axis-0.15,y_xr_auc,markerfmt='skyblue',label="x-ray",linefmt=':')
_,stems_2,_=ax[0].stem(y_ct_auc,'orange')
# ax[0].set_xlabel("Dataset")
ax[0].set_ylabel("AUC",fontsize=16)

# for stem in stems_1:
#     stem.set_linewidth=10

markers_1,stemlines_1, baseline = ax[1].stem(x_axis-0.15, y_xr_acc)
plt.setp(baseline,linestyle=":",linewidth=9)

# ax[1].stem(x_axis-0.15,y_xr_acc,'skyblue',markerfmt='skyblue',linefmt=':')
ax[1].stem(y_ct_acc,'orange',label="CT scans")
# ax[0].set_xlabel("Dataset")
ax[1].set_ylabel("ACC",fontsize=16)

ax[2].stem(x_axis-0.15,y_xr_spec,'g',markerfmt='skyblue',linefmt=':')
ax[2].stem(y_ct_spec,'orange',label="CT scans")
# ax[0].set_xlabel("Dataset")
ax[2].set_ylabel("SPEC",fontsize=16)

ax[3].stem(x_axis-0.15,y_xr_sen,'g',markerfmt='skyblue',linefmt=':',label="x-ray")
ax[3].stem(y_ct_sen,'orange',label="CT scans")
# ax[0].set_xlabel("Dataset")
ax[3].set_ylabel("SEN",fontsize=16)

plt.setp(ax,xticks=[0,1,2,3],xticklabels=x)
plt.tight_layout()

# plt.legend()
plt.show()
```

```python
sns.set_style('darkgrid')
width=0.5
fig,ax=plt.subplots(1,3,figsize=(14,3))
x_axis= np.arange(len(x))
ax[0].stem(x_axis-0.15,y_xr_dunn,markerfmt='skyblue',label="x-ray",linefmt=':')
ax[0].stem(y_ct_dunn,'orange')
# ax[0].set_xlabel("Dataset")
ax[0].set_ylabel("Dunn index",fontsize=16)

ax[1].stem(x_axis-0.15,y_xr_dav,'skyblue',markerfmt='skyblue',linefmt=':')
ax[1].stem(y_ct_dav,'orange',label="CT scans")
# ax[0].set_xlabel("Dataset")
ax[1].set_ylabel("Davies Bouldin",fontsize=16)

ax[2].stem(x_axis-0.15,y_xr_sil,'g',markerfmt='skyblue',linefmt=':', label="X-rays")
ax[2].stem(y_ct_sil,'orange',label="CT scans")
# ax[0].set_xlabel("Dataset")
ax[2].set_ylabel("Silhouette index",fontsize=16)

# ax[3].stem(x_axis-0.15,y_xr_sen,'g',markerfmt='skyblue',linefmt=':',label="x-ray")
# ax[3].stem(y_ct_sen,'orange',label="CT scans")
# # ax[0].set_xlabel("Dataset")
# ax[3].set_ylabel("SEN",fontsize=12)

plt.setp(ax,xticks=[0,1,2,3],xticklabels=x)
plt.tight_layout()
plt.legend(loc='best')
plt.show()
```
